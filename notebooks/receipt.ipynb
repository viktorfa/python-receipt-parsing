```python
overlaps = [[] for _ in full_text_with_y_axis]
overlap_map = [[] for _ in full_text_with_y_axis]
OVERLAP_SAME_LINE_LIMIT = 20

for i, (a_text, (a_1, a_2)) in enumerate(full_text_with_y_axis):
    if i == 0:
        continue
    a_height = a_2 - a_1
    for j, (_, (b_1, b_2)) in enumerate(full_text_with_y_axis):
        b_height = b_2 - b_1
        overlap_size = min(a_1 - b_1 + a_height, b_1 - a_1 + b_height, a_height, b_height)
        overlaps[i].append(max(0, overlap_size))
        if overlap_size >= a_height:
            overlap_map[i].append(j)
            
for i in range(1, 10):
    print(full_text_with_y_axis[i])
    pprint(overlaps[i])
```

```python
OVERLAP_SAME_LINE_LIMIT = 30

class ReceiptData:
    def __init__(self, overlaps, text_annotations):
        self.overlaps = overlaps
        self.text_annotations = text_annotations
        self.token_lines = []
        for o in overlaps:
            self.token_lines.append([text_annotations[i]['description'] for i in o if i != 0])
        self.string_lines = [' '.join(str(x)) for x in self.token_lines]

def read_receipt_from_google_ocr_json(json_dict, OVERLAP_SAME_LINE_LIMIT=30):
    text_annotations = json_dict['textAnnotations']
    full_text_with_y_axis = [(annotation['description'], (annotation['boundingPoly']['vertices'][0]['y'], annotation['boundingPoly']['vertices'][2]['y'])) for annotation in text_annotations]
    overlap_map = get_overlap_map(full_text_with_y_axis)
    distinct_overlaps = get_distinct_lists(overlap_map)
    return ReceiptData(distinct_overlaps, text_annotations)
    
    
def get_overlap_map(full_text_with_y_axis, OVERLAP_SAME_LINE_LIMIT=30):
    result = [[] for _ in full_text_with_y_axis]
    for i, (_, a) in enumerate(full_text_with_y_axis):
        for j, (_, b) in enumerate(full_text_with_y_axis):
            overlap_size = get_overlap_size(a, b)
            if overlap_size >= OVERLAP_SAME_LINE_LIMIT:
                result[i].append(j)
                
    return result
    
def get_height(a):
    return a[1] - a[0]

def get_overlap_size(a, b):
    return min(a[1] - b[0], b[1] - a[0], get_height(a), get_height(b))

def get_distinct_lists(list_of_lists):
    hashable_lists = []
    for l in list_of_lists:
        json_list = json.dumps(l)
        if json_list not in hashable_lists:
            hashable_lists.append(json_list)
    return [json.loads(x) for x in hashable_lists]
    
```
