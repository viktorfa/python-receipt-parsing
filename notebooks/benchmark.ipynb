```python
import re

price_pattern = r'\d+[,\.]\d{2}'
weight_pattern = r'\d+[,\.]\d{3}'
float_pattern = r'\d+[,\.]\d+'
int_pattern = r'(?<![,\.\d+])\d+(?![,\.\d+])'
vat_pattern = r'\d{1,2}\%'
product_text_pattern = r'[A-Za-z]+'
group_price_pattern = r'(' + price_pattern + r')'
group_weight_pattern = r'(' + weight_pattern + r')'
group_float_pattern = r'(' + float_pattern + r')'
group_int_pattern = r'(' + int_pattern + r')'
group_vat_pattern = r'(' + vat_pattern + r')'

compiled_group_price_pattern = re.compile(group_price_pattern)
compiled_group_weight_pattern = re.compile(group_weight_pattern)
compiled_group_float_pattern = re.compile(group_float_pattern)
compiled_group_int_pattern = re.compile(group_int_pattern)
compiled_group_vat_pattern = re.compile(group_vat_pattern)

quantity_units = (
    'g',
    'hg',
    'kg',
    'l',
    'ml',
    'dl',
    'stk',
)

sum_tokens = (
    'sum',
    'total',
    'nok',
)

unit_pattern = r'/?('
for i, x in enumerate(quantity_units):
    unit_pattern += x + r'|' if i < len(quantity_units) - 1 else x
unit_pattern += r')'

compiled_unit_pattern = re.compile(unit_pattern, re.IGNORECASE)

sum_pattern = r'('
for i, x in enumerate(sum_tokens):
    sum_pattern += x + r'|' if i < len(sum_tokens) - 1 else x
sum_pattern += r')'

compiled_sum_pattern = re.compile(sum_pattern, re.IGNORECASE)
```

```python
def get_match_object_from_match(match):
    return dict(
        group=match.group(),
        end_pos=match.end(),
        start_pos=match.start(),
        match=match,
        string=match.string,
    )

def get_matches_from_string_line(line):
    return dict(
        int_matches=[get_match_object_from_match(match) for match in compiled_group_int_pattern.finditer(line)],
        float_matches=[get_match_object_from_match(match) for match in compiled_group_float_pattern.finditer(line)],
        unit_matches=[get_match_object_from_match(match) for match in compiled_unit_pattern.finditer(line)],
    )

def parse_float(string):
    return float(string.replace(',', '.'))

def find_float_type(float_match, unit_matches):
    for unit_match in unit_matches:
        if float_match['end_pos'] == unit_match['start_pos']:
            if unit_match['group'][0] == '/':
                return dict(
                    type='QUANTITY_UNIT_PRICE',
                    value=parse_float(float_match['group']),
                    unit=unit_match['group'][1:],
                    float_match=float_match['match'],
                    unit_match=unit_match['match'],
                )
            else:
                return dict(
                    type='QUANTITY',
                    value=parse_float(float_match['group']),
                    unit=unit_match['group'],
                    float_match=float_match['match'],
                    unit_match=unit_match['match'],
                )
    if is_sum(float_match):
        return dict(
            type='SUM',
            value=parse_float(float_match['group']),
            float_match=float_match['match'],
            unit=None,
            unit_match=None,
        )
    elif is_product_price(float_match):
        return dict(
            type='PRODUCT_PRICE',
            value=parse_float(float_match['group']),
            float_match=float_match['match'],
            unit=None,
            unit_match=None,
        )
    else:
        return dict(
            type='UNKNOWN',
            value=parse_float(float_match['group']),
            float_match=float_match['match'],
            unit=None,
            unit_match=None,
        )

def is_product_price(float_match):
    if not re.match(price_pattern, float_match['group']):
        return False
    words = re.findall(r'\D+', float_match['string'][:float_match['start_pos']])
    if not words:
        return False
    return len(max(words, key=len)) > 2 # Whether there is a word with more the 2 letters in the line

def is_sum(float_match):
    if not re.match(price_pattern, float_match['group']):
        return False
    return compiled_sum_pattern.search(float_match['string'])

def find_int_type(int_match, unit_matches):
    for unit_match in unit_matches:
        if int_match['end_pos'] == unit_match['start_pos']:
            return dict(
                type='QUANTITY',
                value=int(int_match['group']),
                unit=unit_match['group'],
                int_match=int_match['match'],
                unit_match=unit_match['match'],
            )
    if is_vat(int_match):
        return dict(
            type='VAT',
            value=int(int_match['group']),
            int_match=int_match['match'],
            unit=None,
            unit_match=None,
        )
    else:
        return dict(
            type='UNKNOWN',
            value=int(int_match['group']),
            int_match=int_match['match'],
            unit=None,
            unit_match=None,
        )
    
def is_vat(int_match):
    if int_match['group'] not in ('0', '15', '25'):
        return False
    if len(int_match['string']) < int_match['end_pos']:
        return False
    return int_match['string'][int_match['end_pos']] == '%'
```

```python
def parse_token_line(token_line):
    floats = [token for token in token_line if compiled_group_float_pattern.match(token)]
    comma_floats = [token_line[i-1:i+2] for i, x in enumerate(token_line) if len(token_line) > i+1 and (token_line[i] == ',' or token_line[i] == '.') and re.match(r'^\d+$', token_line[i-1]) and re.match(r'^\d+$', token_line[i+1])]
    
    return list(comma_floats)


def remove_comma_floats(token_line):
    new_token_line = list([x for x in token_line])
    parsed_floats = []
    start_pos = 0
    for i, token in enumerate(token_line):
        if i == 0 or i == len(token_line) - 1:
            continue
        if i > 1:
            start_pos += len(token_line[i-2])
        prev_token = token_line[i-1]
        next_token = token_line[i+1]
        if re.match(r'^[,\.]$', token) and re.match(r'^\d+$', prev_token) and re.match(r'^\d+$', next_token):
            value = parse_float(prev_token + token + next_token)
            start_pos = start_pos
            end_pos = start_pos + len(prev_token) + len(token) + len(next_token)
            new_token_line[i-1] = None
            new_token_line[i] = None
            new_token_line[i+1] = None
            parsed_floats.append(dict(
                group=prev_token + token + next_token,
                string=''.join(token_line),
                start_pos=start_pos,
                end_pos=end_pos,
            ))
    return list([x for x in new_token_line if x]), parsed_floats
    
def parse_token_float(token_float):
    return parse_float(''.join(token_float))
```

```python
from recread.parsing.enums import token_types, line_types
from recread.parsing.tokens import get_product_name

def get_line_type(receipt_line):
    occurences = {token_type: 0 for token_type in (token_types.PRODUCT_PRICE, token_types.QUANTITY, token_types.QUANTITY_UNIT_PRICE, token_types.SUM, token_types.UNKNOWN, token_types.VAT_PERCENTAGE)}
    for x in receipt_line.parsed_line:
        occurences[x['type']] += 1
    if occurences[token_types.QUANTITY] and occurences[token_types.QUANTITY_UNIT_PRICE]:
        return line_types.MEASURE
    elif occurences[token_types.SUM]:
        return line_types.TOTAL_SUM
    elif occurences[token_types.PRODUCT_PRICE] and is_product_price(receipt_line.string_line):
        return line_types.PRODUCT
    else: 
        return line_types.UNKNOWN
    
def is_product_price(string, start=0, end=-1):
    # Whether there is a word with more the 2 letters in the line
    pprint(f'{string}, {start}, {end}, {string[start:end]}')
    return len(get_product_name(string[start:end])) > 2 

```

```python
import math

def process_products(receipt):
    pprint(receipt)
    product_candidates = [x for x in receipt.receipt_lines if x.type == line_types.PRODUCT]
    measure_candidates = [x for x in receipt.receipt_lines if x.type == line_types.MEASURE]
    
    pprint(product_candidates)
    pprint(measure_candidates)
    
    for measure in measure_candidates:
        for product in product_candidates:
            if math.isclose(measure.price['value'] * measure.quantity['value'], product.price['value'], abs_tol=.01):
                print("ANUS")
                product.quantity = measure.quantity
                product.unit_price = measure.price
                print(product.quantity)
                
    return receipt
    
    
```
