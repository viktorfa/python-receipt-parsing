```python
import hashlib
import base64

from pprint import pprint

from google.cloud.storage.blob import Blob
from google.cloud import storage

bucket_name = 'vfa-receipt-scanning_receipt-images'



def generate_file_name(image_bytes):
    m = hashlib.sha256()
    m.update(image_bytes)
    return '{}.jpg'.format(m.hexdigest())


def upload_image_from_b64string(b64string, bucket_name=bucket_name):
    storage_client = storage.Client.from_service_account_json(os.path.join(project_dir, '.credentials', 'gcloud-credentials.json'))
    bucket = storage_client.get_bucket(bucket_name)
    
    image_bytes = base64.b64decode(b64string)
    blob = bucket.blob(generate_file_name(image_bytes))
    
    return blob.upload_from_string(image_bytes, content_type='image/jpeg')
    
upload = upload_image_from_b64string(b64image_string)

upload

def get_file_from_bucket(bucket_name, file_name):
    storage_client = storage.Client.from_service_account_json(os.path.join(project_dir, '.credentials', 'gcloud-credentials.json'))
    bucket = storage_client.get_bucket(bucket_name)
    
    blob = bucket.get_blob(file_name)
    print('blob md5hash: {}'.format(blob.md5_hash))
    print('blob id: {}'.format(blob.id))
    
    pprint(blob.metadata)
    
    return blob.download_as_string()


blob_string = get_file_from_bucket(bucket_name, generate_file_name(base64.b64decode(b64image_string)))
blob_string


```
